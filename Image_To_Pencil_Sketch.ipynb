```python
import cv2
from google.colab.patches import cv2_imshow as cvm
import matplotlib.pyplot as plt
from skimage import io
```

```python
src="https://tinyjpg.com/images/social/website.jpg"
```

```python
image=io.imread(src)
```

```python
image
```

```python
cvm(image)
```

```python
image = cv2.resize(image,(500,450))
cvm(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image
```

```python
cvm(gray_image)
```

```python
inverted_image = cv2.bitwise_not(gray_image)
inverted_image
```

```python
cvm(inverted_image)
```

```python
image_smoothing = cv2.GaussianBlur(inverted_image, (21, 21), sigmaX=0, sigmaY=0)
image_smoothing
```

```python
cvm(image_smoothing)
```

```python
def pencil_sketch(x, y):
  return cv2.divide(x, 255 - y, scale=256)
```

```python
pencil_sketch = pencil_sketch(gray_image, image_smoothing)
cvm(pencil_sketch)
```
